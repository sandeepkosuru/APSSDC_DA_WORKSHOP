#### Robust Scalar
- If your dataset contains lot more no of outliers, then Robust Scaling is the best technique for scaling


```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
data.head(5)
```

```python
data.info()
data.shape
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
data.info()
```

```python
#robust sclar formula
#  xi-Q2(x)
# ------------
#  Q3(x)-Q1(x)

###Inter Quartikle Range(IQR)
#Q1 ---->25%(Lower Quaratile)
#Q2 ---->5%(Medium Quaratile)
#Q3 ---->75%(Upper Quaratile)
```

# Apply Robust Scaler Model

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs= RobustScaler()
rs
```

```python
rs2=rs.fit_transform(data)
rs2
```

```python
rs2=pd.DataFrame(rs2,columns= data.columns)
rs2.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, (ax1,ax2)=plt.subplots(ncols = 2,figsize=(9,5))
sns.kdeplot(data['TV'],ax=ax1)
sns.kdeplot(data['radio'],ax=ax1)
sns.kdeplot(data['newspaper'],ax=ax1)
sns.kdeplot(data['sales'],ax=ax1)

sns.kdeplot(rs2['TV'],ax=ax2)
sns.kdeplot(rs2['radio'],ax=ax2)
sns.kdeplot(rs2['newspaper'],ax=ax2)
sns.kdeplot(rs2['sales'],ax=ax2)
```

# Standard Scalar
-most common used scalling technique. it scales between 0 to mean

```python
### Standard Scalr Formula

#  xi=mean(x)
#  ----------
#  std(x)
```

```python
#step:1 import model
#step:2 intialize model into new object
#step:3 fit the model to database
#step:4 convert array value into dataframe
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss
```

```python
ss2=ss.fit_transform(data)
ss2
```

```python
ss2=pd.DataFrame(ss2,columns= data.columns)
ss2.head()
```
